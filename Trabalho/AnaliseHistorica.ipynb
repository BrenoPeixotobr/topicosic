<h1>Análise Histórica do Enade</h1>

<p>As tabelas referentes as edições 2008, 2011, 2014 e 2017 do Enade, disponibilizadas pelo Inep, foram utilizadas para obter a média geral no exame e a média no curso de Ciência da Computação agrupadas por região do Brasil em cada um destes anos. Esses anos foram escolhidos por serem os anos em que os acadêmicos de Ciência da Computação prestam o exame. A tabela com esses resultados é apresentada abaixo e a partir dela foi realizada uma análise para observar como o desempenho geral dos estudantes e do curso de Ciência da Computação evoluiu ao longo desse período.</p>

```python
# Importação das bibliotecas necessárias
import pandas as pd # Manipular dataframes
import matplotlib.pyplot as plt # plotar gráficos
```

```python
enade = pd.read_csv('resultados.csv', '\t') # lendo o .csv, o separador utilizado é o \t
colunas = ['ANO', 'PER_GERAL', 'PER_REGIAO', 'MEDIA_GERAL', 'MEDIA_CC', 'TOTAL_DE_ALUNOS_RE', 'TOTAL_ALUNOS_CC', 'DESVIO_PA', 'TOTAL_ALUNOS', 'REGIAO']
tabela = pd.DataFrame(enade, columns=colunas)


tabela.MEDIA_GERAL = tabela['MEDIA_GERAL'].str.replace(',', '.') #substitui vírgula por ponto
tabela['MEDIA_GERAL'] = pd.to_numeric(tabela['MEDIA_GERAL']) #converte de str para float

tabela.MEDIA_CC = tabela['MEDIA_CC'].str.replace(',', '.') #substitui vírgula por ponto
tabela['MEDIA_CC'] = pd.to_numeric(tabela['MEDIA_CC']) #converte de str para float

tabela.PER_GERAL = tabela['PER_GERAL'].str.replace(',', '.') #substitui vírgula por ponto
tabela['PER_GERAL'] = pd.to_numeric(tabela['PER_GERAL']) #converte de str para float

tabela.PER_REGIAO = tabela['PER_REGIAO'].str.replace(',', '.') #substitui vírgula por ponto
tabela['PER_REGIAO'] = pd.to_numeric(tabela['PER_REGIAO']) #converte de str para float


tabela = tabela.sort_values(by='ANO') # Ordena por ano
tabela.head(21) # Exibindo todas as 21 linhas do arquivo
```

```python
plt.plot(tabela.ANO[tabela.REGIAO == 'N'], tabela.MEDIA_GERAL[tabela.REGIAO == 'N'], label='Norte')
plt.plot(tabela.ANO[tabela.REGIAO == 'S'], tabela.MEDIA_GERAL[tabela.REGIAO == 'S'], label='Sul')
plt.plot(tabela.ANO[tabela.REGIAO == 'SE'], tabela.MEDIA_GERAL[tabela.REGIAO == 'SE'], label='Sudeste')
plt.plot(tabela.ANO[tabela.REGIAO == 'NE'], tabela.MEDIA_GERAL[tabela.REGIAO == 'NE'], label='Nordeste')
plt.plot(tabela.ANO[tabela.REGIAO == 'CO'], tabela.MEDIA_GERAL[tabela.REGIAO == 'CO'], label='Centro-oeste')
plt.legend()
plt.title('Média da nota geral de cada região por ano')
plt.xlabel('Ano')
plt.ylabel('Média da Nota Geral')
```

```python
plt.plot(tabela.ANO[tabela.REGIAO == 'N'], tabela.MEDIA_CC[tabela.REGIAO == 'N'], label='Norte')
plt.plot(tabela.ANO[tabela.REGIAO == 'S'], tabela.MEDIA_CC[tabela.REGIAO == 'S'], label='Sul')
plt.plot(tabela.ANO[tabela.REGIAO == 'SE'], tabela.MEDIA_CC[tabela.REGIAO == 'SE'], label='Sudeste')
plt.plot(tabela.ANO[tabela.REGIAO == 'NE'], tabela.MEDIA_CC[tabela.REGIAO == 'NE'], label='Nordeste')
plt.plot(tabela.ANO[tabela.REGIAO == 'CO'], tabela.MEDIA_CC[tabela.REGIAO == 'CO'], label='Centro-oeste')
plt.legend()
plt.title('Média da nota de computação de cada região por ano')
plt.xlabel('Ano')
plt.ylabel('Média da Nota de Computação')
```
